```python
# Imports PIL module
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# open method used to open different extension image file
im = Image.open(r"lotus.jpg")
```

```python
type(im)
```

```python
# This method will show image in any image viewer
#im.show()
plt.figure(figsize = (15,15))
imgplot = plt.imshow(im)
```

```python
imarray = np.array(im)
```

```python
type(imarray)
```

```python
np.shape(imarray)
```

```python
type(imarray)
```

```python
row_count = np.shape(imarray)[0]
col_count = np.shape(imarray)[1]
color_count = np.shape(imarray)[2]
```

```python
output = np.zeros(np.shape(imarray),dtype=int)
np.shape(output)
```

```python
type(output)
```

```python
for r in range(row_count):
    for c in range(col_count):
        red = int(imarray[r,c,0])
        green =  int(imarray[r,c,1])
        blue =  int(imarray[r,c,2])
        
        avg = (red + green + blue) / 3
        output[r,c,0] = blue
        output[r,c,1] = blue
        output[r,c,2] = blue
        
```

```python
im2 = Image.fromarray(output)
```

```python
#im2.show()
```

```python
plt.figure(figsize = (15,15))
imgplot = plt.imshow(im2)
```
